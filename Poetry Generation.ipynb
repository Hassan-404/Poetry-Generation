Tokenize the sentences of corpus with the help of SpaCy Library

* I make a blank urdu model and pass each and every verse in the spacy object. 
* Object of spacy automatically tokenize the verse word by word. I think it split on the bases of spaces and special characters 

```python
import spacy
unlp = spacy.blank('ur')
# Corpus List will contain all of the verses from ghalib, iqbal and faiz
corpus = list()

# I made iqbal, ghalib, faiz Spacylist for test purposes 
# Otherwise they are not ncessary
iqbal = list()
ghalib = list()
faiz = list()


# Here passage1 , passage2 and passage3 denote text files of iqbal, ghalib and faiz respectively
# In every passage i am building corpus 

# ////////////////passage 1 //////////////////////////////////////////////
for i in passage1:
  if not i:
    continue
  iqbal.append(unlp(i[0]))
  corpus.append(unlp(i[0]))

# ////////////////passage 2 //////////////////////////////////////////////
for i in passage2:
  if not i:
    continue
  ghalib.append(unlp(i[0]))
  corpus.append(unlp(i[0]))

# ////////////////passage 3 //////////////////////////////////////////////
for i in passage3:
  if not i:
    continue
  faiz.append(unlp(i[0]))
```

Generations of Bigrams from corpus

```python
# Here I defined a bigrams list 
bigrams = list()

# Now traversing in the corpus sentence by sentence
for sentence in corpus:
  # In each sentence traversing words
  for word in range(len(sentence)-1):
    # building first word for bigram
    w1 = sentence[word]
    # building second word for bigram
    w2 = sentence[word+1]
    # append the tuple in the list
    bigrams.append(tuple((w1, w2)))
```

```python
# Here I defined a trigrams list 
trigrams = list()

# Now traversing in the corpus sentence by sentence
for sentence in corpus:
  # In each sentence traversing words
  for word in range(len(sentence)-3):
    # building first word for trigram
    w1 = sentence[word]
    # building second word for trigram
    w2 = sentence[word+1]
    # building third word for trigram
    w3 = sentence[word+2]
    # append the tuple in the list
    trigrams.append(tuple((w1, w2, w3)))
```

* Here I have a method **language_model(grams)** which takes grams as an output and find the probabilities of specific grams

**next_lafz(prevLafz)** This method will find next word into the bigram  

```python
# This function contain two lists 
# Words will contain the the tuples of bigrams
# probs will contain the probabilities of particulat gram 
def next_lafz(side, model ,prevLafz):
  prevLafz = unlp(prevLafz)

  #   //// Variables for forward Model \\\\\\\\\\\\\\\\\\
  forwTriWords = list() # It contains the trigram tuple of words
  forwBiWords = list() # it contain the bigram tuple of words
  forwTriProbs = list() # it contains the trigram probability
  forwBiProbs = list() # it contains the bigram probability
  
  #   //// Variables for backward Model \\\\\\\\\\\\\\\\\\
  backTriWords = list() # It contains the trigram tuple of words
  backBiWords = list() # it contain the bigram tuple of words
  backTriProbs = list() # it contains the trigram probability
  backBiProbs = list() # it contains the bigram probability
  
  
  # /////////////// For Forward Models start here

  if side == 'forward':  
    # Search prev Word in Bigrams model
    for next in forwProbListBigram:
      if next[0][0].text == prevLafz.text:
        forwBiWords.append(next[0])
        forwBiProbs.append(next[1])

    # Search prev Word in Trigrams model
    for next in forwProbListTrigram:
      if (next[0][0].text or next[0][1]) == prevLafz.text:
        forwTriWords.append(next[0])
        forwTriProbs.append(next[1])
    
    nextBiLafz = ' '
    nextTriLafz = ' '
    if forwBiWords:
      nextBiLafz = forwBiWords[forwBiProbs.index(max(forwBiProbs))][1]
    if forwTriWords:
      nextTriLafz = forwTriWords[forwTriProbs.index(max(forwTriProbs))][2]
    if model == 'bigram':
      return nextBiLafz
    if model == 'trigram':
      return nextTriLafz
    # //////// Forward Model Ends here
    
   # //////// Backward Model Start here
  elif side == 'backward':

    # Search prev Word in Bigrams model
    for next in backProbListBigram:
      if next[0][1].text == prevLafz.text:
        backBiWords.append(next[0])
        backBiProbs.append(next[1])

    # Search prev Word in Trigrams model
    for next in backProbListTrigram:
      if (next[0][2].text or next[0][1]) == prevLafz.text:
        backTriWords.append(next[0])
        backTriProbs.append(next[1])
      
    nextBiLafz = ' '
    nextTriLafz = ' '
    if backBiWords:
      nextBiLafz = backBiWords[backBiProbs.index(max(backBiProbs))][0]
    if backTriWords:
      nextTriLafz = backTriWords[backTriProbs.index(max(backTriProbs))][0]
    if model == 'bigram':
      return nextBiLafz
    if model == 'trigram':
      return nextTriLafz
      # //////// Backward Model end here
    
```
